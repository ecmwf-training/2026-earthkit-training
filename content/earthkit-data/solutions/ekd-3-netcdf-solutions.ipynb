# earthkit-data: NetCDF

In this notebook you will see how to:

- convert NetCDF to Xarray
- convert NetCDF to a fieldlist

First we read some NetCDF from the remote samples with [from_source](https://earthkit-data.readthedocs.io/en/latest/concepts/inputs/from_source.html). The returned object provides some basic information but its primary goal is to convert the data into the required representation for further work. The actual data loading is deferred as much as possible, until the data is converted into a given type.

```python
import earthkit.data as ekd

d = ekd.from_source("sample", "era5_tquv_pl_subarea.nc")
d
```

```python
d.available_types
```

### Conversion to Xarray

The easiest way to work with NetCDF data is to convert it into Xarray.

```python
d.to_xarray()
```

### Conversion to fieldlist

When the data stucture allows it it is also possible to load NetCDF data into a fieldlist.

```python
fl = d.to_fieldlist()
len(fl)
```
