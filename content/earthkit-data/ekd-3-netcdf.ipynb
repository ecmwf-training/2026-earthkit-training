# Using NetCDF

In this notebook you will see how to:

- convert NetCDF to Xarray
- convert NetCDF to a fieldlist

### Getting the data

First we read some NetCDF from the remote samples with [from_source](https://earthkit-data.readthedocs.io/en/latest/concepts/inputs/from_source.html). The returned object provides some basic information but its primary goal is to convert the data into the required representation for further work. The actual data loading is deferred as much as possible, until the data is converted into a given type.

```python
import earthkit.data as ekd

d = ekd.from_source("sample", "era5_tquv_pl_subarea.nc")
d
```

```python
d.available_types
```

### Conversion to Xarray

The easiest way to work with NetCDF data is to convert it into Xarray.

```python
d.to_xarray()
```

### Conversion to fieldlist

It is also possible to load NetCDF data into a [FieldList](https://earthkit-data.readthedocs.io/en/latest/autoapi/earthkit/data/core/fieldlist/index.html#earthkit.data.core.fieldlist.FieldList), which might be necessary for fieldlist based workflows. This is done by slicing the dataarrays along the relevant dimensions to form the fields.

```python
fl = d.to_fieldlist()
```

The fieldlist contains 48 fields.

```python
len(fl)
```

```python
# first 5 fields
fl.head()
```

A [field](https://earthkit-data.readthedocs.io/en/latest/autoapi/earthkit/data/core/field/index.html#earthkit.data.core.field.Field) generated from NetCDF/Xarray behaves just like as if it was loaded from GRIB (as far as the components and high level metadata is concerned). 

```python
f = fl[0]
f
```

```python
f.get("time.base_datetime")
```

```python
f.time.base_datetime()
```

```python
f.vertical.level_type()
```
