## earthkit-data: BUFR

```python
import earthkit.data as ekd
```

## Getting the data

```python
d = ekd.from_source("file", "synop_10.bufr")
d
```

### Featurelists and BUFR messages

A featuresist is similar to a fieldList, but it is an iterable of "features", where a "feature" can be anything. The best representation of BUFR data is a featurelist where the features are BUFR messages.

```python
fl = d.to_featurelist()
```

```python
len(fl)
```

```python
fl.ls()
```

#### BUFR messages

```python
# f is the first message in the featurelist
f = fl[0]
f
```

```python
f.describe()
```

### Subsetting

```python
fl1 = fl.sel(dataSubCategory=1, ident=[60545, 48352])
fl1.ls()
```

### Converting to pandas

```python
df = fl.to_pandas(columns=["latitude", "longitude",
                           "heightOfStation","airTemperatureAt2M"])
df
```

```python
df = fl.to_pandas(reader="synop")
df
```
